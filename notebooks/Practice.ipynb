```python
from pathlib import Path
from src.preprocessing import preprocess_image
from src.ocr_engine import OCREngine
from src.text_extraction import extract_target_from_ocr
from src.utils import list_images, save_json, draw_highlight
import cv2

images = list_images("../tests")  # or "../data/test_images" from notebooks

ocr = OCREngine()
orig, processed = preprocess_image(images[0])
lines = ocr.run_ocr(processed)
result = extract_target_from_ocr(lines)

out_json = Path("results/json") / (Path(images[0]).name + ".json")
save_json({"image_name": Path(images[0]).name, **result}, str(out_json))

highlighted = draw_highlight(orig, result["target_line"], result["all_lines"])
out_img = Path("results/screenshots") / Path(images[0]).name
out_img.parent.mkdir(parents=True, exist_ok=True)
cv2.imwrite(str(out_img), highlighted)
```
